## Get Polarity and Subjectivity

```python
def get_sentiment(text):
    blob=TextBlob(text)
    return blob.sentiment.polarity, blob.sentiment.subjectivity

df[['Polarity','Subjectivity']]=df['Review'].apply(
    lambda x: pd.Series(get_sentiment(x))
)
df
```

## Convert Polarity to Sentiment Label

```python
def get_label(score):
    if score>0:
        return 'Positive'
    elif score<0:
        return 'Negative'
    else:
        return 'Neutral'

df['Sentiment']=df['Polarity'].apply(get_label)
df
```

```python
movie = input("Enter Movie Name: ")

movie_df = df[df["Movie_Name"].str.lower() == movie.lower()]

total_reviews = len(movie_df)

print("\nMovie Name:", movie)
print("Total Reviews:", total_reviews)

positive = len(movie_df[movie_df["Sentiment"] == "Positive"])
negative = len(movie_df[movie_df["Sentiment"] == "Negative"])
neutral = len(movie_df[movie_df["Sentiment"] == "Neutral"])

print(f"Positive : {positive} ({positive/total_reviews*100:.2f}%)")
print(f"Negative : {negative} ({negative/total_reviews*100:.2f}%)")
print(f"Neutral  : {neutral} ({neutral/total_reviews*100:.2f}%)")
```
