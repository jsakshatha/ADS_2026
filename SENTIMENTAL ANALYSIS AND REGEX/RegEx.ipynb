```python
import re
#!pip install re
```

```python
sen = 'Machine learning is a growing technology which enables computers to learn automatically from past data. We make machine to do specific task without explicitly programming. Machine learning uses various algorithms for building mathematical models and making predictions using historical data or information.'
#sen = 'Machine learning is a growing technology which enables computers to learn \
#automatically from past data. We make machine to do specific task without explicitly\
#programming. Machine learning uses various algorithms for building mathematical models\
#and making predictions using historical data or information.'
```

```python
re.search('Machine',sen)   #Spot the first occurance of the pattern
```

```python
re.search('xyzabc',sen)
```

```python

re.findall('machine',sen, flags=re.IGNORECASE)
```

```python
re.findall('machine',sen)
```

```python
re.split('machine',sen,flags=re.IGNORECASE)
```

```python
re.split(r'\.',sen)
```

```python
re.sub('machine','*',sen,flags=re.IGNORECASE)
```

```python
re.findall(r'\bma.....\b',sen,flags=re.IGNORECASE)     # . indicates any single character, \b indicates a word boundary
```

```python
re.findall(r'\bma[a-z]{5}\b',sen,flags=re.IGNORECASE)   #Search for ma that follows with 5 characters from set a-z
```

```python
re.findall(r'\bma[a-z]+\b',sen,flags=re.IGNORECASE)   #Search for ma that follows with any number of lowercase alphabets
```

```python
re.findall(r'ma[a-z]+',sen,flags=re.IGNORECASE)   #Search for ma. Taking out word boundary will search for pattern in the middle of text too.
```

```python
re.findall(r'\bm.chin.',sen, flags=re.IGNORECASE)
```

```python
sen = 'ma Ma mac machine bmac cmacd macd'
#re.findall(r'\b[a-z]{3,5}\b',sen)
re.findall(r'\b[a-z]{4}\b',sen)
```

```python
sen = 'She is 20. She learnt driving when she was 18. I only taught her.'
pattern = r'\b[A-Za-z]+\b'
#Try pattern as ., [a-z],[a-z]*,[a-z]+,[a-zA-Z0-9]+,[a-zA-Z0-9]*. Each time try putting \b in starting and ending of pattern
re.findall(pattern,sen)
```

```python
#Find all words that start with vowels. Looks for a word with atleast one character after the vowel.
sen = 'a an the are articles'
re.findall(r'\b[aeiou][a-z]*',sen)
```

```python
re.findall(r'\b[aeiou][a-z]+',sen)
```

```python
#Find all words that start with vowels. Lets 'a' alone be counted
re.findall(r'\b[aeiou][a-z]*',sen)
```

```python
sen = 'I have list of email ids here. url@connect.com, apple@berry.com, web@browser.com. @Navin, @Pravin, please trace the mails.'
#Find the people who are tagged in this message
print(re.findall(r' @([A-Za-z]+)',sen))


#Find the email ids listed in the message
print(re.findall(r'[a-zA-Z_0-9]+@[a-z]+\.com',sen))

#Check if the message ends with 'mails.'
print(re.findall('mails.$',sen))

#Check if the message starts with I or We
print(re.findall('^(I|We)',sen))   #^ indicates beginning of sentence and | indicates 'or'
```
